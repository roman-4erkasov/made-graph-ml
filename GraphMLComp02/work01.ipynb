```python
def get_degrees(nodes, graph):
  degree_map = nx.degree(graph)
  degrees = np.array([degree_map[n] for n in nodes])
  return degrees

def get_probabilities(nodes, graph):
    degrees = get_degrees(nodes, graph)
    p = degrees/degrees.sum()
    return p

p = get_probabilities(graph.nodes(), graph)
np.random.choice(graph.nodes(),size=10, p=p)

strategy = []
available = set(np.random.choice(graph.nodes(),size=10, p=p))
for _ in range(60):
  p = get_probabilities(graph.nodes(), graph)
  chosen = np.random.choice(graph.nodes(),size=10, p=p).tolist()
  available = available - set(chosen)
  strategy.append(chosen)
```
